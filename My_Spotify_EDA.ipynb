# Setup and Initial Exploration

```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Load JSON files
audio1=pd.read_json('Streaming_History_Audio_2019-2022_0.json')
audio2=pd.read_json('Streaming_History_Audio_2022-2024_1.json')
video=pd.read_json('Streaming_History_Video_2021-2024.json')
```

# Data Cleaning and Preparation

```python
#Convert ts to Datetime:
audio1['ts'] = pd.to_datetime(audio1['ts'])
audio2['ts'] = pd.to_datetime(audio2['ts'])
video['ts'] = pd.to_datetime(video['ts'])
```

```python
#Drop or Impute Missing Values:
# Drop columns with more than 90% missing values
audio1 = audio1.dropna(axis=1, thresh=0.1 * len(audio1))
audio2 = audio2.dropna(axis=1, thresh=0.1 * len(audio2))
video = video.dropna(axis=1, thresh=0.1 * len(video))

    
```

```python
#Impute missing values for user_agent_decrypted with "Unknown".
audio1['user_agent_decrypted'].fillna('Unknown', inplace=True)
audio2['user_agent_decrypted'].fillna('Unknown', inplace=True)
```

# Handle Inconsistent Data:

```python
# Drop duplicate rows
audio1 = audio1.drop_duplicates()
audio2 = audio2.drop_duplicates()
video = video.drop_duplicates()
```

```python
# Concatenate audio and video data
combined = pd.concat([audio1, audio2, video], ignore_index=True)
```

```python
combined['platform'] = combined['platform'].str.lower()
```

```python
combined.isnull().sum()
```

```python
# Drop rows where more than 50% of the columns have null values
combined = combined.dropna(thresh=0.5 * len(combined.columns))
```

```python
combined.info()
```

# Data Analysis

### 1. Overall Listening Time

```python
# Convert ms_played to hours
combined['hours_played'] = combined['ms_played'] / (1000 * 60 * 60)
total_hours = combined['hours_played'].sum()
print(f'Total listening time: {total_hours:.2f} hours')
```

### 2. Listening Time by Platform

```python
# Group by platform and sum up hours played
platform_hours = combined.groupby('platform')['hours_played'].sum().reset_index()
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(y='platform', x='hours_played', data=platform_hours)  # Swap x and y
plt.title('Listening Time by Platform')
plt.xlabel('Hours Played')
plt.ylabel('Platform')
plt.show()
```

### 3. Top Tracks and Artists

```python
# Group by track and artist, then sum up hours played
top_tracks = combined.groupby('master_metadata_track_name')['hours_played'].sum().reset_index()
top_artists = combined.groupby('master_metadata_album_artist_name')['hours_played'].sum().reset_index()

# Get top 10 tracks and artists
top_tracks = top_tracks.sort_values(by='hours_played', ascending=False).head(10)
top_artists = top_artists.sort_values(by='hours_played', ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='hours_played', y='master_metadata_track_name', data=top_tracks)
plt.title('Top 10 Tracks by Listening Time')
plt.xlabel('Hours Played')
plt.ylabel('Track Name')
plt.show()

plt.figure(figsize=(12, 8))
sns.barplot(x='hours_played', y='master_metadata_album_artist_name', data=top_artists)
plt.title('Top 10 Artists by Listening Time')
plt.xlabel('Hours Played')
plt.ylabel('Artist Name')
plt.show()
```

### 4. What are the peak listening hours throughout the day?

```python
# Extract hour from timestamp
combined['hour'] = combined['ts'].dt.hour

# Aggregate listening time by hour
hourly_listening = combined.groupby('hour')['hours_played'].sum().reset_index()

# Plot the listening hours
plt.figure(figsize=(12, 6))
sns.lineplot(x='hour', y='hours_played', data=hourly_listening, marker='o')
plt.title('Total Listening Time by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Hours Played')
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()
```

### 5. How does your listening behavior differ between weekdays and weekends?

```python
# Extract day of the week from timestamp (0=Monday, 6=Sunday)
combined['day_of_week'] = combined['ts'].dt.dayofweek

# Classify days into weekdays (0-4) and weekends (5-6)
combined['day_type'] = combined['day_of_week'].apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')

# Aggregate listening time by day type
day_type_listening = combined.groupby('day_type')['hours_played'].sum().reset_index()

# Plot the listening time by day type
plt.figure(figsize=(8, 6))
sns.barplot(x='day_type', y='hours_played', data=day_type_listening)
plt.title('Listening Time: Weekdays vs Weekends')
plt.xlabel('Day Type')
plt.ylabel('Total Hours Played')
plt.show()
```

### 6. What percentage of your listening time is spent on shuffled versus non-shuffled tracks?

```python
# Aggregate listening time by shuffle mode
shuffle_distribution = combined.groupby('shuffle')['hours_played'].sum().reset_index()
shuffle_distribution['percentage'] = (shuffle_distribution['hours_played'] / shuffle_distribution['hours_played'].sum()) * 100

# Plot the distribution of shuffle mode
plt.figure(figsize=(8, 6))
sns.barplot(x='shuffle', y='percentage', data=shuffle_distribution, palette='pastel')
plt.title('Percentage of Listening Time: Shuffled vs Non-Shuffled')
plt.xlabel('Shuffle Mode')
plt.ylabel('Percentage of Total Listening Time')
plt.show()
```

### 7. Analyze Listening Trends Over Time

```python
# Extract month and year from timestamp
combined['month'] = combined['ts'].dt.to_period('M')

# Aggregate listening time by month
monthly_trends = combined.groupby('month')['hours_played'].sum().reset_index()
monthly_trends['month'] = monthly_trends['month'].astype(str)  # Convert Period to string for better plotting

# Plot the monthly listening trends
plt.figure(figsize=(16, 7))
sns.lineplot(x='month', y='hours_played', data=monthly_trends, marker='o', color='teal')
plt.title('Monthly Listening Trends')
plt.xlabel('Month')
plt.ylabel('Total Hours Played')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### 8. Discover Listening Patterns for Different Days of the Week

```python
# Aggregate listening time by day of the week
weekly_listening = combined.groupby('day_of_week')['hours_played'].sum().reset_index()
weekly_listening['day_of_week'] = weekly_listening['day_of_week'].apply(lambda x: ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'][x])

# Plot the listening time by day of the week
plt.figure(figsize=(10, 6))
sns.barplot(x='day_of_week', y='hours_played', data=weekly_listening, palette='coolwarm')
plt.title('Listening Time by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Hours Played')
plt.xticks(rotation=45)
plt.show()
```

### 9. Most Frequent Reasons for Starting and Stopping Playback

```python
# Count occurrences of each reason for starting playback
start_reasons = combined['reason_start'].value_counts().reset_index()
start_reasons.columns = ['reason', 'count']

# Plot reasons for starting playback
plt.figure(figsize=(12, 6))
sns.barplot(x='count', y='reason', data=start_reasons, palette='pastel')
plt.title('Reasons for Starting Playback')
plt.xlabel('Count')
plt.ylabel('Reason')
plt.show()

# Count occurrences of each reason for stopping playback
end_reasons = combined['reason_end'].value_counts().reset_index()
end_reasons.columns = ['reason', 'count']

# Plot reasons for stopping playback
plt.figure(figsize=(12, 6))
sns.barplot(x='count', y='reason', data=end_reasons, palette='pastel')
plt.title('Reasons for Stopping Playback')
plt.xlabel('Count')
plt.ylabel('Reason')
plt.show()
```
